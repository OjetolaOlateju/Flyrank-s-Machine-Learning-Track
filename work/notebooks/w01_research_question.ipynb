I'm choosing Lane 4: CTR / Engagement Opportunity Scoring.

From Discovery A, I found that search volume has almost no relationship 
with how often a page actually gets shown (impressions) — correlation of 0.001. 
This means visibility alone doesn't explain page performance.

From Discovery B, I found CTR drops sharply as position gets worse (35.5% at

But when I checked pages within the SAME position tier, I found their CTR

determine engagement — something about the page itself matters too.

I want to focus on this gap: pages that are already visible, but 
underperforming on engagement relative to similar pages, since that's a 
realistic, fixable opportunity (rewriting titles/metadata) rather than 
chasing visibility alone.

## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

Research question: Once a page is visible in search results, why do some 
pages convert that visibility into strong engagement (clicks) while others 
at the same position level do not — and which underperforming pages should 
be prioritized for review?

Unit of analysis: one page.

Decision this informs: which visible-but-underperforming pages the content 
team should review first for a title/metadata rewrite.

Action: rewrite the page's title or meta description, or improve intent 
match, for pages flagged as low-CTR relative to their position tier.

Cost of a wrong call: wasted content-team effort rewriting a page whose 
real problem isn't the title/metadata but something else (e.g. weak 
position, low relevance, seasonal decline) — meaning the rewrite wouldn't 
actually help.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

# Number 1: search volume barely predicts impressions
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")

# Number 2: CTR drops sharply by position tier
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print("\nMean CTR by position tier:")
print(ctr_by_pos.round(4).to_string())

# Number 3: CTR still varies a lot WITHIN the same position tier
same_tier = df[(df["position_tier"] == "striking") & (df["impressions_90d"] >= 100)]
print(f"\nCTR spread within 'striking' tier (n={same_tier.shape[0]}):")
print(same_tier["ctr"].describe())
```

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

This project can say: which visible pages show unusually low engagement relative to similar pages (same position tier), based on observed data — and can rank those pages as review candidates.

This project cannot say: that a title/metadata rewrite will definitely fix engagement (that would require an experiment, not just this data); that we've found a Google ranking factor; or that any single page's low CTR is definitely caused by its title rather than something else (seasonality, consolidation, or noise). All findings are observed and directional, not proof of causation.
